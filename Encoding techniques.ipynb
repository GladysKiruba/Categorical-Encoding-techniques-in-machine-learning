## 1) Label Encoder

```python
# Importing Label encoder
from sklearn import preprocessing

# Creating object
LE=preprocessing.LabelEncoder()
```

```python
# Encoding the Grades column
Data['Grades']=LE.fit_transform(Data['Grades'])

# Uniques values in the grade column after encoding
Data['Grades'].unique()
```

```python
# Grades column after encoding
Data['Grades'].head(10)
```

## 2) One Hot Encoder

```python
# One hot encoder with dummies
# Getting dummies for the column city
Dummies=pd.get_dummies(Data.City)
Dummies
```

```python
# Joining the dummies with the dataset
JoinedData=pd.concat([Data,Dummies],axis='columns')
JoinedData
```

```python
# Dropping city column and one of the dummy column
FinalData=JoinedData.drop(['City','Pune'],axis= 'columns')
FinalData
```

```python
# Importing One hot encoder from sklearn
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
```

```python
# Encoding the gender column
Data['Gender'] = OHE.fit_transform(Data.Gender.values.reshape(-1, 1)).toarray()
Data['Gender'].unique()
```

```python
# Printing first five encoded column
Data['Gender'].head()
```

```python
# Joining encoded gender column with dataset
NewData=pd.concat([Data,Data['Gender']],axis='columns')
NewData.head(10)
```
