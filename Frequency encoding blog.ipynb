```python
# Frequency distribution
FreqDist=Data.groupby('City').size()/len(Data)
FreqDist
```

```python
# Creating new column by mapping the frequency and the city column
DataFreq['FreqOfCity']=DataFreq.City.map(FreqDist)
DataFreq['FreqOfCity']
```
